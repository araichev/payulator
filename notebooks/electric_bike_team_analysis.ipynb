```python
from pathlib import Path

import pandas as pd
import numpy as np


DATA_DIR = Path('../loans_active/EBT-20180916/misc')
```

```python
# Read and tidy shop expenses
#
path = DATA_DIR/"Electric Bike Team Limited - Profit and Loss to Sep 2018.xlsx"

# Transpose
f = pd.read_excel(path).T[1:].reset_index()
f.columns = ['date', 'stock_expenses', 'other_expenses', 'sales']

# Drop nan rows
f = f.dropna(subset=['stock_expenses', 'other_expenses', 'sales'])

# Correct dates
new_dates = pd.date_range('2016-04-01', '2017-06-30', freq='M')
new_dates
f['date'] = new_dates
f = f.set_index('date')

def clean(x):
    y = str(x).replace('$', '').replace(',', '')
    return float(y)

f = f.applymap(clean)
f
```

```python
# Compute profit
f['expenses'] = f['stock_expenses'] + f['other_expenses']
f['net_profit'] = f['sales'] - f['expenses']
f['net_profit_ratio'] = f['net_profit']/f['sales']
f
```

```python
# Compute some stats
#

g = f.sum()
n = f.shape[0]
g['net_profit_ratio'] = g['net_profit']/g['sales']
g['monthly_sales'] = g['sales']/n
g['monthly_net_profit'] = g['net_profit']/n
g
```
